```python
import pandas as pd
import pymc3 as pm
from IPython.display import Image
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

## Section 16.1

```python
data = pd.read_csv("data/TwoGroupIQ.csv")
smart_drug = data[data["Group"] == "Smart Drug"]
placebo = data[data["Group"] == "Placebo"]
print(data.groupby("Group").agg(["mean", "std"]))
Image("images/fig16_2.png")
```

```python
with pm.Model() as model16_1:
    obs_sd = smart_drug["Score"].std()
    mu = pm.Normal("mu", mu=smart_drug["Score"].mean(), sd=obs_sd*100)
    sd = pm.Uniform("sd", obs_sd / 1000, obs_sd * 1000)
    obs = pm.Normal("obs", mu=mu,sd=sd, observed=smart_drug["Score"])
    
    traces = pm.sample(10000)
```

```python
burned_trace = traces[500:]
fig, [(ax1, ax2), (ax3, ax4)] = plt.subplots(2,2, figsize=(10,6))

#Posterior Mean
pm.plot_posterior(burned_trace['mu'], point_estimate='mode', ref_val=100, ax=ax1)
ax1.set_xlabel('$\mu$')
ax1.set_title('Mean')

#Data with credible posterior mu's and std's superimposed - posterior predictive check


tr_len = len(burned_trace)
curves = 20
skip_distance = int(len(burned_trace) / curves)
indexes = np.arange(0, len(burned_trace), skip_distance)
means = burned_trace["mu"][indexes]
std_devs = burned_trace["sd"][indexes]

x = np.arange(smart_drug["Score"].min() - 20, smart_drug["Score"].max())
x = np.tile(x.reshape(-1,1), (1,curves))
posterior_estimates = scipy.stats.norm.pdf(x, means , std_devs)
ax2.plot(x, posterior_estimates)
ax2.hist(smart_drug["Score"], density=True, bins=25)

[ax2.spines[spine].set_visible(False) for spine in ['left', 'right', 'top']]


#bottom left Plot standard deviation posterior
pm.plot_posterior(burned_trace["sd"], point_estimate='mode', ax=ax3)
ax3.set_xlabel('$\sigma$')
ax3.set_title('Standard deviation')


#bottom right, effect size posterior
pm.plot_posterior((burned_trace['mu']-100)/burned_trace['sd'], point_estimate='mode', ref_val=0,ax=ax4)
ax4.set_title('Effect Size')
ax4.set_xlabel('$(\mu - 100)/\sigma$')
```

# Helpers

```python
def plot_posterior(title, x_label, axes, traces):
    pm.plot_posterior(traces, point_estimate='mode', ref_val=100, ax=axes)
    axes.set_xlabel(x_label)
    axes.set_title(title)
```

```python
def posterior_predictive_check(curves, burned_trace, axes):
    tr_len = len(burned_trace)
    skip_distance = int(len(burned_trace) / curves)
    indexes = np.arange(0, len(burned_trace), skip_distance)
    means = burned_trace['mu'][indexes]
    std_devs = burned_trace['sd'][indexes]

    x = np.arange(smart_drug["Score"].min() - 20, smart_drug["Score"].max())
    x = np.tile(x.reshape(-1,1), (1,curves))
    posterior_estimates = scipy.stats.norm.pdf(x, means , std_devs)
    axes.plot(x, posterior_estimates)
    axes.hist(smart_drug["Score"], density=True, bins=25)
    axes.set_title("Data with Posterior Predictive")

    [axes.spines[spine].set_visible(False) for spine in ['left', 'right', 'top']]
```
